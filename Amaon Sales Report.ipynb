```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
Data = pd.read_csv(r'Amazon Sale Report.csv')
```

```python
Data.drop(['New'],axis=1,inplace=True)
```

```python
Data.drop(['PendingS'],axis=1,inplace=True)
```

```python
# sum of null valueabs
pd.isnull(Data).sum()
```

```python
Data.dropna(inplace=True) #drop null value
```

```python
Data.info()
```

```python
Data.columns
```

```python
# change datatypeabs
Data['ship-postal-code'] = Data['ship-postal-code'].astype('int')
```

```python
Data.info()
```

```python
# check Datatype 
Data['ship-postal-code'].dtype
```

```python
Data['Date'] = pd.to_datetime(Data['Date'])
```

```python
Data.info()
```

```python
# Rename Column 
Data.rename(columns={'Qty':' Quantity'},inplace=True)
```

```python
Data.columns
```

```python
ax=sns.countplot(x='Size',data=Data,color="green")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Note In the above Graph we can see that most people buy M-size 
```

```python
Data.groupby(['Size'],as_index= False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
S_Qty = Data.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=S_Qty)
```

```python
# Histogram
```

```python
Data['Size'].hist()
```

```python
Data['Category']=Data['Category'].astype(str)
Column_data = Data['Category']
plt.figure(figsize=(10,5))
plt.hist(Column_data,bins=30,edgecolor="red",color="purple")
plt.xticks(rotation=90)
plt.show()
```

```python
#From Above Graph we can see that Most of the buyers are T-shirt
```

```python
#check B2B data by using pie chart 
b2b_check = Data['B2B'].value_counts()
plt.pie(b2b_check,labels=b2b_check.index,autopct='%1.1f%%')
plt.show()
```

# From above Chart We can see that maximum that 99.3% of buyers are retailers and 0.8% are B2B buyers 

```python
x_label=Data['Category']
y_label=Data['Size']
plt.scatter(x_label,y_label)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#plot count of cities by state 
plt.figure(figsize=(10,5))
sns.countplot(data=Data,x='ship-state',color="red")
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('distribution of state')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top 10 States
Top_10_state = Data['ship-state'].value_counts().head(10)
#plot count by states
plt.figure(figsize=(12,6))
sns.countplot(data=Data[Data['ship-state'].isin(Top_10_state.index)],x='ship-state',color="yellow")
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('distribution of state')
plt.xticks(rotation=90)
plt.show()
```
